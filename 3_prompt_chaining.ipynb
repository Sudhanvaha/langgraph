```python
from langgraph.graph import StateGraph,START,END
from langchain_huggingface import HuggingFaceEndpoint,ChatHuggingFace
from typing import TypedDict
from dotenv import load_dotenv
load_dotenv()

llm=HuggingFaceEndpoint(
    repo_id="deepseek-ai/DeepSeek-V3.2-Exp",
    task="text-generation"
)
model=ChatHuggingFace(llm=llm)
```

```python
class BlogState(TypedDict):
    title:str
    outline:str
    content:str
```

```python
def create_outline(state:BlogState)->BlogState:
    state['outline']=model.invoke(f"generate a outline for {state['title']}").content
    return state
def create_blog(state:BlogState)->BlogState:
    state['content']=model.invoke(f"generate a blog for this {state['title']} within this outline {state['outline']}")
    return state
```

```python
graph=StateGraph(BlogState)

graph.add_node("create_outline",create_outline)
graph.add_node("create_blog",create_blog)


graph.add_edge(START,"create_outline")
graph.add_edge("create_outline","create_blog")
graph.add_edge("create_blog",END)

workflow=graph.compile()

```

```python
initial_state={"title":"why everone tells All terrorists are muslims but all muslims are not terrorist"}

final_state=workflow.invoke(initial_state)
print(final_state)
```
